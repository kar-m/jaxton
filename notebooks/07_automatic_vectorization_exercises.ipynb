```python
import jax
import jax.numpy as jnp

jax.devices()
```

```python
## sample batch data
score_data = jnp.array([[3., 1., 2.], [6., 2., 4.], [9., 3., 6.]])
weights = jnp.array([0.2, 0.3, 0.5])
points_a = jnp.array([[0., 0.], [1., 1.], [2., 0.]])
points_b = jnp.array([[1., 0.], [2., 2.]])

score_data
```

```python
## sample scoring functions
def normalize_row(row):
    return row / jnp.sum(row)

def weighted_score(row, weights):
    return jnp.sum(row * weights)

def distance(point_a, point_b):
    return jnp.sqrt(jnp.sum((point_a - point_b) ** 2))
```

**Exercise 61: Vectorize `normalize_row` over the rows of `score_data` and assign it to `normalize_rows`**

```python
normalize_rows = jax.vmap(normalize_row)
```

**Exercise 62: Run `normalize_rows` on `score_data` and assign it to `normalized_scores`**

```python
normalized_scores = normalize_rows(score_data)
normalized_scores
```

**Exercise 63: Vectorize `weighted_score` over rows while sharing `weights` and assign it to `batched_weighted_score`**

```python
batched_weighted_score = jax.vmap(weighted_score, in_axes=(0, None))
```

**Exercise 64: Run `batched_weighted_score` on `score_data` and `weights` and assign it to `weighted_scores`**

```python
weighted_scores = batched_weighted_score(score_data, weights)
weighted_scores
```

**Exercise 65: Vectorize `distance` over `points_b` for a single point in `points_a` and assign it to `distances_to_b`**

```python
distances_to_b = jax.vmap(distance, in_axes=(None, 0))
```

**Exercise 66: Run `distances_to_b` with the first point of `points_a` and assign it to `first_point_distances`**

```python
first_point_distances = distances_to_b(points_a[0], points_b)
first_point_distances
```

**Exercise 67: Create a pairwise distance function for `points_a` and `points_b` using nested vectorization and assign it to `pairwise_distances`**

```python
pairwise_distances = jax.vmap(distances_to_b, in_axes=(0, None))
```

**Exercise 68: Run `pairwise_distances` on `points_a` and `points_b` and assign it to `distance_matrix`**

```python
pairwise_distances(points_a, points_b)
```

**Exercise 69: Vectorize `normalize_row` over rows and place the mapped axis at axis=1, assigning it to `normalize_rows_columns`**

```python
normalize_rows_columns = jax.vmap(normalize_row, in_axes=(0), out_axes=(1))
```

**Exercise 70: Run `normalize_rows_columns` on `score_data` and assign it to `normalized_scores_columns`**

```python
normalized_scores_columns = normalize_rows_columns(score_data)
normalized_scores_columns
```
